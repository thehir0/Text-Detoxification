```python
import pandas as pd
detox_data = pd.read_csv('../data/raw/filtered.tsv', sep='\t', index_col=0)
detox_data
```

```python
detox_data.info()
```

```python
detox_data = detox_data[detox_data['ref_tox'] >= detox_data['trn_tox']]
detox_data = detox_data.drop(columns=['similarity', 'lenght_diff', 'ref_tox', 'trn_tox'])
detox_data = detox_data.reset_index(drop=True)
detox_data
```

```python
paradetox_data = pd.read_csv('../data/external/paradetox.tsv', sep='\t')

paradetox_data.head(5)
```

```python
paradetox_data = paradetox_data.drop(columns=paradetox_data.columns.values[2:])
paradetox_data.columns = detox_data.columns.values
paradetox_data
```

```python
data = pd.concat([detox_data, paradetox_data], ignore_index=True)
data
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
train_data
```

```python
test_data
```

```python
train_data.to_csv('../data/interim/train_filtered.csv')
test_data.to_csv('../data/interim/test_filtered.csv')
```
